# Bitcoin Price Prediction

```python
import numpy as np
import pandas as pd
from sklearn import metrics
import seaborn as sns
```

```python
train = pd.read_csv("TRAIN.csv")
test = pd.read_csv("TEST.csv")
```

```python
train
```

```python
test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train["Date"] = le.fit_transform(train["Date"])
```

```python
train
```

```python
train.drop(["Adj Close","Volume"],axis=1,inplace=True)
```

```python
train
```

```python
test["Date"] = le.fit_transform(test["Date"])
```

```python
test
```

```python
x = train.iloc[:,0:-1].values
y = train.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.shape
```

```python
x_test
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(test)
```

```python
y_pred
```

```python
val = [[898,9335.47,9348.37,9004.90]]
pred = reg.predict(val)
print(pred)
```

```python
print('train accuracy: ', reg.score(x_train,y_train),'\ntest accuracy: ', reg.score(x_test,y_test) )
```

```python
Output = pd.DataFrame({"Date" : pd.read_csv("TEST.csv")["Date"], "Close" : y_pred})
```
